```python
import requests
import pandas as pd

# Define the API endpoint and parameters
stem = "https://api.census.gov/data/timeseries/intltrade/"
database = "exports/hs"
endpoint = stem + database


Years = ["2013","2014","2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022"]
Months = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
Hs_Codes = ["3915100000", "3915200000", "3915300000", "3915900010", "3915900090", "4707100000", "4707200020", "4707200040", "4707300020", "4707300040", "4707900000", "5103100000", "5103200000","5103300000","5104000000","5202100000","5202910000","5202990000","5505100020","5505100040","5505200000","6310100000","6310900000","7204100000","7204210000","7204290000",
            "7204300000","7204410020","7204410040","7204410060","7204410080","7204490020","7204490040","7204490060","7204490070","7204490080","7204500000"]




```

```python
for year in Years:
    for month in Months:
        for hs in Hs_Codes:
            params = {
                "get": "E_COMMODITY,UNIT_QY1,UNIT_QY2,CTY_CODE,CTY_NAME,DISTRICT,DIST_NAME,YEAR,MONTH,CC_MO,QTY_1_MO,ALL_VAL_MO,AIR_VAL_MO,AIR_WGT_MO,VES_VAL_MO,VES_WGT_MO,CNT_VAL_MO,CNT_WGT_MO",
                "YEAR": year,
                "MONTH": month,
                "E_COMMODITY": "3915100000"
            }

            # Print the complete URL for debugging
            print(f"Request URL: {endpoint}")
            print(f"Parameters: {params}")

            # Make the request to the API
            response = requests.get(endpoint, params=params)

            # Check if the request was successful
            if response.status_code == 200:
                # Print the response content for debugging
                print("Response Content:", response.content)
                
                # Parse the JSON response
                try:
                    data = response.json()
                    # Check the structure of the data
                    print("Parsed Data:", data)
                    
                    # Convert the data to a pandas DataFrame
                    if len(data) > 1:  # Ensure there's data to process
                        df = pd.DataFrame(data[1:], columns=data[0])
                        # Display the DataFrame
                        print(df)
                        
                        # Save the DataFrame to a CSV file
                        save_stem = "C:/Users/hmc98/Documents/GitHub/Wasteonomics/"
                        
                        # Saving on different folders wrt 4 digit hs codes
                        if "3915" in hs:
                            file = "3915/"
                        elif "4707" in hs:
                            file = "4707/"
                        elif "5103" in hs:
                            file = "5103/"
                        elif "5104" in hs:
                            file = "5104/"
                        elif "5202" in hs:
                            file = "5202/"
                        elif "5505" in hs:
                            file = "5505/"
                        elif "6310" in hs:
                            file = "6310/"
                        elif "7204" in hs:
                            file = "7204/"

                        save_stem_hs = save_stem + file
                        hs_year_month =  hs + "_" + year + "_" + month
                        filename = save_stem_hs + hs_year_month + ".csv"

                        df.to_csv(filename, index=False)
                    else:
                        print("No data available for the given parameters.")
                except ValueError as e:
                    print(f"Error parsing JSON: {e}")
            else:
                print(f"Error: {response.status_code}")
                print("Response Content:", response.content)
    
```
